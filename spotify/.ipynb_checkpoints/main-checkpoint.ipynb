### Spotify API

```python
import os
from dotenv import load_dotenv
import urllib.parse
import requests
import spotipy
from spotipy.oauth2 import SpotifyOAuth
import pandas as pd
import base64
import json
import webbrowser

# Wczytanie zmiennych środowiskowych z pliku .env
load_dotenv()

client_id = os.getenv("SPOTIPY_CLIENT_ID")
client_secret = os.getenv("SPOTIPY_CLIENT_SECRET")
redirect_url = "http://localhost:8888/callback"
auth_url = "https://accounts.spotify.com/authorize"
token_url = "https://accounts.spotify.com/api/token"
api_base_url = "https://api.spotify.com/v1"

def get_login():
    """
    Funkcja generuje URL do autoryzacji i otwiera go w przeglądarce.
    """
    scope = 'user-read-private user-read-email'

    params = {
        "client_id": client_id,
        "response_type": "code",
        "redirect_uri": redirect_url,
        "scope": scope,
        "show_dialog": True
    }

    url = f"{auth_url}?{urllib.parse.urlencode(params)}"
    webbrowser.open(url)  # Otwarcie URL w przeglądarce

    return url

def get_access_token(code):
    """
    Funkcja wymienia kod autoryzacji na token dostępu.
    """
    auth = base64.b64encode(f"{client_id}:{client_secret}".encode("utf-8")).decode("utf-8")

    headers = {
        "Authorization": f"Basic {auth}",
        "Content-Type": "application/x-www-form-urlencoded"
    }

    data = {
        "grant_type": "authorization_code",
        "code": code,
        "redirect_uri": redirect_url
    }

    response = requests.post(token_url, headers=headers, data=data)

    if response.status_code == 200:
        token_data = response.json()
        access_token = token_data['access_token']
        refresh_token = token_data['refresh_token']
        print(f"Access Token: {access_token}")
        print(f"Refresh Token: {refresh_token}")
        return access_token, refresh_token
    else:
        print(f"Error: {response.status_code}")
        return None, None

def get_user_profile(access_token):
    """
    Funkcja do pobrania profilu użytkownika za pomocą tokena dostępu.
    """
    headers = {
        "Authorization": f"Bearer {access_token}"
    }

    user_url = f"{api_base_url}/me"
    response = requests.get(user_url, headers=headers)

    if response.status_code == 200:
        user_data = response.json()
        return user_data
    else:
        print(f"Error: {response.status_code}")
        return None

def refresh_access_token(refresh_token):
    """
    Funkcja do odświeżenia tokenu dostępu za pomocą refresh_token.
    """
    auth = base64.b64encode(f"{client_id}:{client_secret}".encode("utf-8")).decode("utf-8")

    headers = {
        "Authorization": f"Basic {auth}",
        "Content-Type": "application/x-www-form-urlencoded"
    }

    data = {
        "grant_type": "refresh_token",
        "refresh_token": refresh_token
    }

    response = requests.post(token_url, headers=headers, data=data)

    if response.status_code == 200:
        token_data = response.json()
        new_access_token = token_data['access_token']
        print(f"New Access Token: {new_access_token}")
        return new_access_token
    else:
        print(f"Error: {response.status_code}")
        return None

# Main flow
if __name__ == "__main__":
    # 1. Wygeneruj URL autoryzacji
    print("Przekieruj użytkownika na ten URL, aby uzyskać kod autoryzacji:")
    print(get_login())  # To otworzy przeglądarkę dla użytkownika

    # 2. Po uzyskaniu kodu z przekierowania użytkownika, wywołaj poniższą funkcję, aby uzyskać token
    code = input("Wprowadź kod autoryzacji (code) z URL przekierowania: ")  # Wprowadź kod ręcznie
    access_token, refresh_token = get_access_token(code)

    if access_token:
        # 3. Pobierz profil użytkownika
        user_profile = get_user_profile(access_token)
        if user_profile:
            print("Dane użytkownika:")
            print(json.dumps(user_profile, indent=4))

        # 4. Opcjonalnie, jeśli token wygasł, odśwież token
        refresh_option = input("Czy chcesz odświeżyć token? (tak/nie): ")
        if refresh_option.lower() == "tak" and refresh_token:
            new_access_token = refresh_access_token(refresh_token)
            if new_access_token:
                # Możesz teraz wykonać zapytania z nowym access_token
                pass
    else:
        print("Nie udało się uzyskać dostępu.")
```

```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth

# Parametry autoryzacji
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(
    client_id=client_id,  # Twoje ID aplikacji
    client_secret=client_secret,  # Twoje sekretne hasło
    redirect_uri="http://localhost:8888/callback",  # Twoja URI przekierowania
    scope="user-top-read"  # Zakres umożliwiający dostęp do topowych utworów
))

# Sprawdź, czy token działa, i użyj go do dalszych zapytań
token = sp.auth_manager.get_access_token()

# Pokaż token
print("Access Token:", token)
```
